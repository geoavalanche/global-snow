```python
st = ulmo.ncdc.gsod.get_stations(country='CZ')
st.keys()
stat_one = st.keys()[2]
ts = ulmo.ncdc.gsod.get_data(stat_one,start='2010-01-01',end='2014-06-30')
```

```python
ints = ts[ts.keys()[0]]
```

```python
dates = []
vals = []
for itm in ints:
    dates.append(itm['date'])
    vals.append(itm['snow_depth'])
```

```python
for idx, v in enumerate(vals):
    if vals[idx] > 100.0:
        vals[idx] = 0
```

```python
plot(dates, vals)
```
